```python
import pandas as pd
import json
import os
import pprint

from urllib.request import urlopen

import matplotlib.pyplot as plt

import plotly.express as px


import dash
import dash_html_components as html

from tqdm import notebook

import plotly.offline as pyo

pp = pprint.PrettyPrinter()
```

```python
original_data = pd.read_csv("DE1_0_2008_Beneficiary_Summary_File_Sample_1.csv")
original_data.shape
```

```python
original_data["sex"] = original_data.BENE_SEX_IDENT_CD.replace(
    [1, 2], ["Male", "Female"]
)
original_data["race"] = original_data.BENE_RACE_CD.replace(
    [1, 2, 3, 5], ["White", "Black", "Other", "Hispanic"]
)

original_data[
    [col for col in original_data.columns if "SP_" == str(col[:3])]
] = original_data[
    [col for col in original_data.columns if "SP_" == str(col[:3])]
].replace(
    2, 0
)


original_data["state"] = original_data.SP_STATE_CODE.replace(
    list(range(1, 55)),
    [
        "AL",
        "AK",
        "AZ",
        "AR",
        "CA",
        "CO",
        "CT",
        "DE",
        "DC",
        "FL",
        "GA",
        "HI",
        "ID",
        "IL",
        "IN",
        "IA",
        "KS",
        "KY",
        "LA",
        "ME",
        "MD",
        "MA",
        "MI",
        "MN",
        "MS",
        "MO",
        "MT",
        "NE",
        "NV",
        "NH",
        "NJ",
        "NM",
        "NY",
        "NC",
        "ND",
        "OH",
        "OK",
        "OR",
        "PA",
        "",
        "RI",
        "SC",
        "SD",
        "TN",
        "TX",
        "UT",
        "VT",
        "",
        "VA",
        "WA",
        "WV",
        "WI",
        "WY",
        "Other",
    ],
)

original_data["ssa"] = original_data.apply(
    lambda row: str(row["SP_STATE_CODE"]).zfill(2)
    + str(row["BENE_COUNTY_CD"]).zfill(3),
    axis=1,
)


payment_columns = [
    "MEDREIMB_IP",
    "BENRES_IP",
    "PPPYMT_IP",
    "MEDREIMB_OP",
    "BENRES_OP",
    "PPPYMT_OP",
    "MEDREIMB_CAR",
    "BENRES_CAR",
    "PPPYMT_CAR",
]
original_data["Total Expense"] = original_data[payment_columns].sum(axis=1)
```

```python
original_data
```

```python
ssa_to_fips = pd.read_excel("geo_info.xlsx")
ssa_to_fips["ssacd"] = ssa_to_fips["ssacd"].apply(lambda x: str(x).zfill(5))
ssa_to_fips["fipscd"] = ssa_to_fips["fipscd"].apply(lambda x: str(x).zfill(5))

ssa_to_fips.head()
```

```python
df = pd.merge(original_data, ssa_to_fips, left_on="ssa", right_on="ssacd")
df.shape
```

```python
df.head()
```

```python
# df.to_excel("Synpuf/clean.xlsx", index=False)
```

```python
condition_cols = [
    col for col in df.columns if "SP_" == str(col[:3]) and "STATE" not in col
]

melted_df = pd.melt(
    df,
    id_vars=[
        "DESYNPUF_ID",
        "fipscd",
        "sex",
        "race",
        "county",
        "state_y",
        "Total Expense",
    ],
    value_vars=condition_cols,
    var_name="Field",
    value_name="Count",
)

conditions_by_county = pd.DataFrame(
    melted_df.groupby(["fipscd", "county", "state_y", "sex", "race", "Field"]).agg(
        {"Count": [sum, len], "Total Expense": sum}
    )
)

conditions_by_county.reset_index(inplace=True)

conditions_by_county.columns = [
    "fipscd",
    "County",
    "State",
    "Sex",
    "Race",
    "Field",
    "Cases",
    "Patients",
    "Total Expense",
]

conditions_by_county_all = pd.DataFrame(
    melted_df.groupby(["fipscd", "county", "state_y", "sex", "race"]).agg(
        {"Count": [sum, len], "Total Expense": sum}
    )
).reset_index()
conditions_by_county_all["Field"] = "All"
conditions_by_county_all.columns = [
    "fipscd",
    "County",
    "State",
    "Sex",
    "Race",
    "Cases",
    "Patients",
    "Total Expense",
    "Field",
]
conditions_by_county_all = conditions_by_county_all[conditions_by_county.columns]


conditions_by_county = pd.concat([conditions_by_county_all, conditions_by_county])


conditions_by_county.head()
```
